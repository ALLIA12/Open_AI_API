```python
from openai import OpenAI
import os
from dotenv import load_dotenv
import pdfplumber
import pandas as pd
from pathlib import Path
from IPython.display import Audio
import sounddevice as sd
import numpy as np
from scipy.io.wavfile import write
```

```python
# Load environment variables from .env file
load_dotenv()

client = OpenAI(
    # api_key=os.getenv('OPENAI_API_KEY_ALI')
    api_key=os.getenv('OPENAI_API_KEY')
)
```

```python
response = client.audio.speech.create(
    model="tts-1",
    voice="fable",
    input="Jeddah is a port city in Makkah Province, Saudi Arabia, located along the Red Sea coast in the Hejaz region",
)
```

```python
# Save the response content to an audio file
speech_file_path = "speech.mp3"
with open(speech_file_path, "wb") as f:
    f.write(response.content)
```

```python
# Display the audio file in the notebook
Audio(speech_file_path)
```

```python
# Function to record audio
def record_audio(duration=5, fs=44100):
    print("Recording...")
    myrecording = sd.rec(int(duration * fs), samplerate=fs, channels=2)
    sd.wait()  # Wait until recording is finished
    print("Done recording.")
    return myrecording
```

```python
# Save the recording to a WAV file
def save_recording(recording, fs=44100, filename='my_recording.wav'):
    write(filename, fs, recording)  # Save as WAV file 
```

```python
# Record for a few seconds
fs = 44100  # Sample rate
duration = 5  # Duration in seconds
```

```python
audio_data = record_audio(duration, fs)
# Save the recording
audio_filename = 'my_recording.wav'
save_recording(audio_data, fs, audio_filename)
#مدينة جِدَّة هي مركز محافظة جدة إحدى محافظات منطقة مكة المكرمة
```

```python
audio_file = open("my_recording.wav", "rb")
translation = client.audio.translations.create(
    model="whisper-1",
    file=audio_file,
)
print(translation.text)
```

```python
audio_file = open("my_recording.wav", "rb")
transcription = client.audio.transcriptions.create(
    model="whisper-1",
    file=audio_file,
    language='ar'
)
print(transcription.text)
```
